```python
df_pitt = df[df["city"]=="Pittsburgh"]
```

```python
df_vegas = df[df["city"]=="Las Vegas"]
```

```python
df_vegas.head()
```

```python
pitt_stars = df_pitt["stars"]
vegas_stars = df_vegas["stars"]
type(pitt_stars)
```

```python
plt.hist(
    pitt_stars,
    alpha=0.3,
    color="yellow",
    label="Pittsburgh",
    bins="auto"
)

plt.hist(
    vegas_stars,
    alpha=0.3,
    color="blue",
    label="Las Vegas",
    bins="auto"
)

plt.xlabel("Rating")
plt.ylabel("Number of Rating scores")

plt.legend(loc = "best")
plt.title("Review Distribution of pittsburgh and Las Vegas")

plt.show()
```

```python
plt.hist(
    [pitt_stars,vegas_stars],
    alpha=0.7,
    color=["red","green"],
    label=["Pittsburgh","Las Vegas"],
    bins = "auto"
)

plt.xlabel("Rating")
plt.ylabel("NUmber of Rating scores")

plt.legend(loc = "upper left")
plt.title("Review Distribution")
```

```python
df_health = df[df["category_0"] == "Health & Medical"]
df_fast = df[df["category_0"]=="Fast Food"]
df_break = df[df["category_0"]=="Breakfast & Brunch"]
df_break.head()
```

```python
plt.scatter(
    df_health["stars"],df_health["review_count"],
    marker = "o",
    color = "r",
    alpha = 0.7,
    s = 124,
    label = ["Health and Medical"]
)

plt.scatter(
    df_fast["stars"],df_fast["review_count"],
    marker = "h",
    color = "b",
    alpha = 0.7,
    s = 124,
    label = ["Fast Food"]
)

plt.scatter(
    df_break["stars"],df_break["review_count"],
    marker = "*",
    color = "g",
    alpha = 0.7,
    s = 124,
    label = ["Breakfast & Brunch"]
)

plt.xlabel("Rating")
plt.ylabel("Review Count")

plt.legend(loc = "best")
axes = plt.gca()
#gca is get current axes
axes.set_yscale("log")
plt.show()
```
